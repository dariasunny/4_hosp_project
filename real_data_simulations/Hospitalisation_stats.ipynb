```python
import pandas as pd
from scipy.stats import poisson
import numpy as np
import seaborn as sns
import random
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
df = pd.read_csv('/content/initial_data_patients.csv', on_bad_lines='skip', header= 0)
```

```python
df = df[df['Возраст'] < 101]
```

```python
df
```

```python
# Density Plot and Histogram of all arrival delays
fig, ax = plt.subplots()
sns.distplot(df['Час'], hist=True, kde=False, color = 'royalblue', hist_kws={'edgecolor':'black'}, ax=ax)
# Plot formatting
plt.title("Histogram of Hours of Patient's Calls")
plt.xlabel('Hour')
ax.set_xticks(range(0,23,2))
plt.ylabel('Density')
```

```python
# Density Plot and Histogram of all arrival delays

fig, ax = plt.subplots()
sns.distplot(df['Возраст'], hist=True, kde=False, color = 'royalblue', hist_kws={'edgecolor':'black'})
# Plot formatting
plt.title("Histogram of Patient's Ages")
plt.xlabel('Age')
ax.set_xticks(range(20,100,4))
plt.ylabel('Density')
```

```python
# Density Plot and Histogram of all arrival delays
fig, ax = plt.subplots()
sns.distplot(df['Разница_полная'], hist=True, kde=True, color = 'royalblue', hist_kws={'edgecolor':'black'}, ax=ax)
# Plot formatting
plt.title('Density Plot and Histogram of Intervals between Calls')
plt.xlabel('Minutes')
plt.ylabel('Density')
ax.set_xlim(1, 1400)
plt.show()
```
