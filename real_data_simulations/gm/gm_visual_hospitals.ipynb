```python
import pandas as pd
from scipy.stats import poisson
import numpy as np
import seaborn as sns
import random
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('.../processed_ok_all_restore_ext.csv', on_bad_lines='skip',sep = ";", header= None)
```

```python
data.columns = ['index', 'adress', 'lat', 'lon', 'final']
```

```python
data
```

```python
data_filtered_lat = data[data['lat'] <= 60.05]
```

```python
data_filtered_lat = data_filtered_lat[data_filtered_lat['lat'] >= 59.8]
```

```python
 data_filtered = data_filtered_lat[data_filtered_lat['lon'] < 30.6]
```

```python
 data_filtered = data_filtered[data_filtered['lon'] >= 30.1]
```

```python
coordinates_filtered = data_filtered[['lat', 'lon']].to_numpy()
```

```python
len(coordinates_filtered)
```

```python
array_coord = coordinates_filtered
  
column_values = ['lat', 'lon']
  
coordinates_filtered_df = pd.DataFrame(data = array_coord, 
                  columns = column_values)
```

```python
coordinates_filtered
```

```python
hosp_1 = [59.849212, 30.295701] #26 boln 
hosp_2 = [60.000713, 30.199033] #mhs
hosp_3 = [60.036281, 30.360164] 
hosp_4 = [59.879137, 30.468124]
```

```python
array = np.array([hosp_1, hosp_2,
                  hosp_3, hosp_4])
column_values = ['lat', 'lon']
hospitals = pd.DataFrame(data = array, 
                  columns = column_values)
```

```python
hospitals
```

```python
hospitals["color"] = 2
```

```python
hospitals["size"] = 16
```

```python
fig = px.scatter_mapbox(coordinates_filtered_df, 
                        lat="lat", 
                        lon="lon", 
                        #color="counts",
                        #color_continuous_scale=color_scale,
                        #size="counts",
                        zoom=8, 
                        height=800,
                        width=800)
fig2 = px.scatter_mapbox(hospitals, lat="lat", lon="lon",zoom=8, color = hospitals["color"], size = hospitals["size"])
fig.add_trace(fig2.data[0]) # adds the line trace to the first figure
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
